```r
install.packages("ISLR")
library(ISLR)
```

```r
set.seed(1)
x = rnorm(100) # normal random values, i.e, between 0-1
hist(x)
```

```r
tmp = rnorm(200,3,1)
hist(tmp)
```

```r
eps = rnorm(100, 0, sqrt(0.05))
```

```r
y = -1 + 0.5*x + eps
```

```r
plot(x,y)
```

```r
# help(lm)
```

```r
lm.fit = lm(y~x) # linear model relate y and x 
summary(lm.fit)
lm.fit
```

```r
plot(x, y)
abline(lm.fit, lwd=3, col=2) #Extract the slope and intercept and slope and uses it
# abline(-1, 0.5, lwd=3, col=3)
abline(mean(y),0,lwd=3,col=4)
legend(-1, legend = c("model fit", "pop. regression","Mean line"), col=c(2,3,4), lwd=3)
```

```r
lm.fit_sq = lm(y~x+I(x^2))
summary(lm.fit_sq)
```

```r
set.seed(1)
eps1 = rnorm(100, 0, 0.125)
x1 = rnorm(100)
y1 = -1 + 0.5*x1 + eps1
plot(x1, y1)
lm.fit1 = lm(y1~x1)
summary(lm.fit1)
```
