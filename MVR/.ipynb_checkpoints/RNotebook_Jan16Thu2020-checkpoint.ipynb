```r
data(mtcars)
class(mtcars)
dim(mtcars)
head(mtcars)
```

```r
X = as.matrix(mtcars)
```

```r
apply(X,2,sd)
```

```r
colMeans(X)
```

```r
Xc = X-rep(colMeans(X), rep.int(nrow(X), ncol(X)))
```

```r
apply(Xc,2,mean)
```

```r
Xc = apply(X, 2, function(cols) cols-mean(cols))
```

```r
apply(Xc,2,mean)
```
